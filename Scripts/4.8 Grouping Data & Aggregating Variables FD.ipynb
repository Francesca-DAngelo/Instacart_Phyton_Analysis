Q4. Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() 
and loc() functions.

```python
#Create a loyalty flag using loc
df_ords_prods_merged['max_order'] = df_ords_prods_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_ords_prods_merged.head(5)
```

```python
# Deriving Columns with loc()
df_ords_prods_merged.loc[df_ords_prods_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merged.loc[(df_ords_prods_merged['max_order'] <= 40) & (df_ords_prods_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
df_ords_prods_merged['loyalty_flag'].value_counts(dropna = False)
```

```python
df_ords_prods_merged[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

Q5. The marketing team at Instacart wants to know whether there’s a difference between the spending habits of the three types of customers you identified. What you’re trying to determine is whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers.

```python
# Examine all the descriptive statistics
df_ords_prods_merged.groupby('loyalty_flag').agg({'prices': ['min', 'max','mean', 'sum']})
```

```python
#Create an average price flag using loc
df_ords_prods_merged['average_price'] = df_ords_prods_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df_ords_prods_merged.head()
```

```python
# Deriving Columns with loc()
df_ords_prods_merged.loc[df_ords_prods_merged['average_price'] < 10, 'average_order_flag'] = 'Low spender'
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['average_price'] >= 10, 'average_order_flag'] = 'High spender'
```

```python
# Checking values
df_ords_prods_merged['average_order_flag'].value_counts(dropna = False)
```

Q7. In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column.

```python
#Create a median prior order flag using loc
df_ords_prods_merged['median_prior_order'] = df_ords_prods_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.mean)
```

```python
# Deriving Columns with loc()
df_ords_prods_merged.loc[df_ords_prods_merged['median_prior_order'] > 20, 'frequency_order_flag'] = 'Non-frequent customer'
```

```python
df_ords_prods_merged.loc[(df_ords_prods_merged['median_prior_order'] > 10) & (df_ords_prods_merged['median_prior_order'] <= 20), 'frequency_order_flag'] = 'Regular customer'
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['median_prior_order'] <= 10, 'frequency_order_flag'] = 'Frequent customer'
```

```python
# Checking values
df_ords_prods_merged['frequency_order_flag'].value_counts(dropna = False)
```
