```python
# Address the duplicates using an appropriate method.
Missing data can be handled in a few different ways:

1. Create a new variable based on the missing value that acts as a flag.
2. Replace the value for the column's mean or median (if the variable is numeric).
3. Remove the missing data.
In this case, I decided to create another column called "new customer" that contains either True or False values to dinstinguish
between old and new customers.
```
